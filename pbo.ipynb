```python
# 1. Mendefinisikan Class Dosen
class Dosen:
    # 2. Atribut nama, nidn, dan mata_kuliah
    def __init__(self, nama, nidn, mata_kuliah):
        self.nama = nama
        self.nidn = nidn
        self.mata_kuliah = mata_kuliah

    # 3. Method mengajar()
    def mengajar(self):
        print(f"Dosen {self.nama} dengan NIDN {self.nidn} sedang mengajar mata kuliah {self.mata_kuliah}.")

# 4. Melakukan instansiasi untuk membuat dua objek dosen yang berbeda
dosen1 = Dosen("Dr. Budi Santoso", "0012097801", "Pemrograman Python")
dosen2 = Dosen("Prof. Ani Setiawati", "0005058502", "Basis Data")

# 5. Memanggil method mengajar() untuk masing-masing objek
print("--- Aktivitas Mengajar ---")
dosen1.mengajar()
dosen2.mengajar()
```
